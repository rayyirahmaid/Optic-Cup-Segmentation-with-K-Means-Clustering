```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from glob import glob
```

```python
# Setup Folder
input_folder = '/content/drive/MyDrive/Tugas_Besar/Dataset_RGB_Retina_Full/training'
histo_red_dir = '/content/drive/MyDrive/Tugas_Besar/Histogram/RED'
histo_green_dir = '/content/drive/MyDrive/Tugas_Besar/Histogram/GREEN'
histo_blue_dir = '/content/drive/MyDrive/Tugas_Besar/Histogram/BLUE'
```

```python
# Buat folder jika belum ada
for d in [histo_red_dir, histo_green_dir, histo_blue_dir]:
    os.makedirs(d, exist_ok=True)

# Dapatkan semua file png
image_files = glob(os.path.join(input_folder, '*.png'))

if not image_files:
    print('Tidak ada file gambar dalam folder ini.')
else:
    for full_file_name in image_files:
        base_file_name = os.path.basename(full_file_name)
        name, ext = os.path.splitext(base_file_name)

        # Baca gambar (OpenCV membaca dalam BGR, kita ubah ke RGB)
        img_bgr = cv2.imread(full_file_name)
        if img_bgr is None: continue
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

        # Pisahkan channel RGB
        R, G, B = cv2.split(img_rgb)

        # Fungsi bantuan untuk plot dan simpan histogram
        def save_histogram(channel_data, color, title, filename):
            plt.figure()
            # Hitung histogram, abaikan nilai 0-5 (seperti di kode MATLAB Anda)
            hist = cv2.calcHist([channel_data], [0], None, [256], [0, 256])
            hist[0:6] = 0

            plt.bar(range(256), hist.flatten(), color=color)
            plt.title(title)
            plt.xlabel('Pixel Intensity')
            plt.ylabel('Frequency')
            plt.grid(True)
            plt.savefig(filename)
            plt.close()

        # Simpan Histogram
        save_histogram(R, 'r', f'Histogram Red for {base_file_name}', os.path.join(histo_red_dir, f'{name}_Red.png'))
        save_histogram(G, 'g', f'Histogram Green for {base_file_name}', os.path.join(histo_green_dir, f'{name}_Green.png'))
        save_histogram(B, 'b', f'Histogram Blue for {base_file_name}', os.path.join(histo_blue_dir, f'{name}_Blue.png'))

print("Proses ekstraksi dan histogram selesai!")
```

```python
import cv2
import matplotlib.pyplot as plt
import os
from glob import glob
import numpy as np
```

```python
# Folder Path
folder1 = '/content/drive/MyDrive/Tugas_Besar/Histogram_Masked_OD/BLUE'
folder2 = '/content/drive/MyDrive/Tugas_Besar/Histogram_Masked_OC/BLUE'
output_folder = '/content/drive/MyDrive/Tugas_Besar/Overlayed_Histogram_OD_OC/BLUE'

os.makedirs(output_folder, exist_ok=True)

image_files1 = sorted(glob(os.path.join(folder1, '*.png')))
image_files2 = sorted(glob(os.path.join(folder2, '*.png')))
```

```python
for file1, file2 in zip(image_files1, image_files2):
    img1 = cv2.imread(file1, cv2.IMREAD_GRAYSCALE)
    img2 = cv2.imread(file2, cv2.IMREAD_GRAYSCALE)

    if img1 is None or img2 is None: continue

    # Hitung histogram
    hist1 = cv2.calcHist([img1], [0], None, [256], [0, 256]).flatten()
    hist2 = cv2.calcHist([img2], [0], None, [256], [0, 256]).flatten()

    # Normalisasi histogram
    if np.sum(hist1) > 0: hist1 = hist1 / np.sum(hist1)
    if np.sum(hist2) > 0: hist2 = hist2 / np.sum(hist2)

    # Plot histogram
    plt.figure()
    plt.plot(hist1, color='r', label='RED OD')
    plt.plot(hist2, color='b', label='RED OC')

    base_name = os.path.splitext(os.path.basename(file1))[0]
    plt.title(f'Overlay Histogram of {base_name}')
    plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))

    output_filename = os.path.join(output_folder, f'{base_name}_OC_Overlayed.png')
    plt.savefig(output_filename, bbox_inches='tight')
    plt.close()

print("Done!")
```
